```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.Species.unique()
```

### Subsets

```python
setosa_sub = df[df.Species=='Iris-setosa']
setosa_sub.head()
```

```python
versicolor_sub = df[df.Species=='Iris-versicolor']
versicolor_sub.head()
```

```python
virginica_sub = df[df.Species=='Iris-virginica']
virginica_sub.head()
```

### Sort data 

```python
df.sort_values('PetalLengthCm')
```

### Merge subsets

```python
merged_data = pd.merge(versicolor_sub,versicolor_sub)
merged_data
```

### Transpose

```python
df.transpose()
```

### Melting 

```python
melted_df = pd.melt(df, id_vars=['Species'], value_vars=['SepalLengthCm','SepalWidthCm'])
```

```python
melted_df
```

```python
pivot_df = pd.pivot_table(df,values=['SepalWidthCm','SepalLengthCm'], index='Species', aggfunc="mean")
```
